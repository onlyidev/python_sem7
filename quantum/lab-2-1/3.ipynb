```python
import pandas as pd
from pandas import DataFrame as df
import itertools
import numpy as np
```

```python
s = np.array([0,0,1,1,1], dtype=np.byte)
def truthTable(s):
    comb = list(itertools.product(np.array([0,1], dtype=np.byte), repeat=len(s)))
    cols = ["s", "x", "f(x)"]
    table = df()
    table[cols[1]] = [str(a) for a in comb]
    table[cols[0]] = str(s)
    table[cols[2]] = [np.dot(a,s)%2 for a in comb]
    return table
truthTable(s)
```

```python
def bernsteinVazirani(bits):
    s = np.random.choice([0,1], size=bits)
    print(f"s={s}")
    return lambda x: np.dot([np.byte(c) for c in x], s) % 2
```

```python
f = bernsteinVazirani(5)
```

```python
def checkIfBernsteinVazirani(outputs):
    sLength = np.int8(np.log2(len(outputs)))
    assert 2**sLength == len(outputs)
    s = np.array([np.byte(outputs[2**i]) for i in range(0,sLength)], dtype=np.byte)
    s = np.flip(s)
    print(f"s={s}")
    table = truthTable(s)
    table["given"] = [np.byte(c) for c in outputs]
    print(table)
    return all(table["f(x)"] == table["given"])
```

```python
checkIfBernsteinVazirani("0001")
```
