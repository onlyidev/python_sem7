```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import Derivative, dsolve, Function
import sympy as sp
from sympy.abc import x
from matplotlib.lines import Line2D
```

```python
y = Function('y')
result = dsolve((x**2*y(x) + 2*y(x))*Derivative(y(x), x) - 3*x*y(x)**2 - x, y(x))
for r in result:
    r.simplify()
    display(r)
```

```python
C2 = sp.symbols("C2")
msol = sp.sqrt((C2*(x**2+2)**3 - 1)/3)
sp.solve(sp.Eq(result[1].rhs, msol), C2)
sp.print_latex(result[1])
```

```python
range=(-4,4)
yrange=(-8,8)

x = np.linspace(*range, 1000)
y = np.linspace(*yrange, 1000)

X, Y = np.meshgrid(x,y)
F = lambda C: C*(X**2+2)**3-1-3*Y**2

contour1 = plt.contour(X, Y, F(0.125), [0], colors='red')
contour2 = plt.contour(X, Y, F(0.05), [0], colors='green')
contour3 = plt.contour(X, Y, F(0.3), [0], colors='blue')

legend_lines = [
    Line2D([0], [0], color='red', lw=2, label='$C=0.125$'),
    Line2D([0], [0], color='green', lw=2, label='$C=0.05$'),
    Line2D([0], [0], color='blue', lw=2, label='$C=0.3$')
]

plt.title("$y = \pm\sqrt{\\frac{C(x^2+2)^3-1}{3}}$")
plt.legend(handles=legend_lines, loc='upper center', bbox_to_anchor=(0.5, 0.3))

plt.show()

```
