```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
import math
```

```python
def dydx(x,y):
    return y**2/x**2 + 7*y/x + 9
```

```python
range = (10,20)
x_pts = np.linspace(*range, 1000)
result = integrate.solve_ivp(dydx, range, [-20, -25, -30], t_eval=x_pts)
result
```

```python
[plt.plot(result.t, ya, label = l) for ya, l in zip(result.y, ["$y(10)=$-20", "$y(10)=$-25", "$y(10)=$-30"])]
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
```

```python
def f(x, C):
    return -x/(np.log(np.abs(C*x)))-3*x
[plt.plot(result.t, [f(t, c) for t in result.t], label=f"C={c}") for c in [0.04, 0.02, 0.001]]
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("$y=-\\frac{x}{\ln{|Cx|}}-3x$")
plt.legend()
plt.show()
```
