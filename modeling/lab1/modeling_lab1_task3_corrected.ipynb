```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import Derivative, dsolve, Function
import sympy as sp
from sympy.abc import x
```

```python
y = Function('y')
result = dsolve(Derivative(y(x), x) + y(x)*sp.tan(x) - 1/sp.cos(x), y(x))
result
```

```python
def f(x, C):
    return np.sin(x) + C*np.cos(x)
arr = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.plot(arr, [f(t, np.sqrt(2)-1) for t in arr], label="$C=\sqrt{2}-1$")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("$y=\sin{x} + C\cos{x}$")
plt.plot(np.pi/4, 1, 'bo', label="Koši taškas")
plt.legend()
plt.show()
```
