```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
import math
```

```python
def dydx(x,y):
    return (3*x*(y**2) + x) / ((x**2)*y + 2*y)
```

```python
range = (0.5,2)
x_pts = np.linspace(*range, 1000)
result = integrate.solve_ivp(dydx, range, [-0.1, 0.1, 0.2, -0.2], t_eval=x_pts)
result
```

```python
plt.plot(result.t, result.y[0,:], label = "y(0.5)=-0.1")
plt.plot(result.t, result.y[1,:], label = "y(0.5)=0.1")
plt.plot(result.t, result.y[2,:], label = "y(0.5)=0.2")
plt.plot(result.t, result.y[3,:], label = "y(0.5)=-0.2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
yrange=(-3,3)
x = np.linspace(*range, 1000)
y = np.linspace(*yrange, 1000)
X, Y = np.meshgrid(x,y)
F = 0.1*(X**2+2)**3-1-3*Y**2
c = plt.contour(X,Y,F,[0])
plt.xlabel("x")
plt.ylabel("y")
plt.title("$y = \pm\sqrt{\\frac{C(x^2+2)^3-1}{3}}$, $C=0.1$")
plt.show()
```
