```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy import Function
from sympy.abc import t
```

```python
t0=0
dx0=3
```

```python
x = Function('x')
eq = sp.Eq(sp.diff(x(t), t, 2) + 4*x(t), 4*sp.sin(4*t) - 4*sp.cos(2.02*t))
sol = sp.dsolve(eq)
sol
```

```python

# x = sp.Symbol("C_1")*sp.cos(2*t) + sp.Symbol("C_2")*sp.sin(2*t) - sp.Rational(1,3)*sp.sin(4*t) + sp.Rational(4,5)*sp.cos(3*t)
# x = sp.Symbol("C_1")*sp.cos(2*t) + sp.Symbol("C_2")*sp.sin(2*t) - sp.Rational(1,3)*sp.sin(2*t) + sp.Rational(4,5)*sp.cos(2*t)
x = sol.rhs
dx = x.diff(t, 1)
eq = [sp.Eq(x.subs(t, 0), t0), sp.Eq(dx.subs(t,0), dx0)]
cauchy = x.subs(sp.solve(eq))
tex = sp.latex(cauchy)

scale=100
# arr = np.linspace(-2*np.pi*scale, 2*scale*np.pi, 10*max(100,scale))
arr = np.linspace(0, 500, 10*max(100,scale))
plt.plot(arr, [cauchy.subs(t, v).evalf(6) for v in arr], label="$x(t)$")
plt.plot([t0-1,t0,t0+1], [cauchy.diff(t).subs(t,t0)*x for x in [t0-1,t0,t0+1]], '--', label=f"$x'({t0})={cauchy.diff(t).subs(t,t0)}$", )
plt.xlabel("t")
plt.ylabel("x")
plt.title(f"$x(t)={tex}$")
plt.plot(0, 0, 'bo', label="Koi takas")
plt.legend()
plt.show()
```
