```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy import Function
from sympy.abc import t
```

```python
x = Function('x')
eq = sp.Eq(sp.diff(x(t), t, 2) + 4*x(t), 4*sp.sin(4*t) - 4*sp.cos(3*t))
sol = sp.dsolve(eq)
sol
```

```python

x = sp.Symbol("C_1")*sp.cos(2*t) + sp.Symbol("C_2")*sp.sin(2*t) - sp.Rational(1,3)*sp.sin(4*t) + sp.Rational(4,5)*sp.cos(3*t)
dx = x.diff(t, 1)
dx.subs(t, 0)
eq = [sp.Eq(x.subs(t, 0), 0), sp.Eq(dx.subs(t,0), 3)]
cauchy = x.subs(sp.solve(eq))
tex = sp.latex(cauchy)

arr = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.plot(arr, [cauchy.subs(t, v).evalf(6) for v in arr], label="$x(t)$")
plt.plot([-1,0,1], [3*x for x in [-1,0,1]], '--', label="$x'(0)=3$", )
plt.xlabel("t")
plt.ylabel("x")
plt.title(f"$x(t)={tex}$")
plt.plot(0, 0, 'bo', label="Koi takas")
plt.legend()
plt.show()
```
